```python
# We search the following strings to determine the cpu usage of this function.["upfd","smfd","amfd","udmd","pcfd","udrd","ausfd","bsfd","nrfd","nssfd","mongo"]
# Each file contains the results of multiple top commands, which are then averaged 
l=["upfd","smfd","amfd","udmd","pcfd","udrd","ausfd","bsfd","nrfd","nssfd","mongo"]
cpu=[[]for i in range(len(l))]
memory=[[]for i in range(len(l))]
cpu_f=[]
mem_f=[]
sys_cpu=[]
x=[10,20,30,40,50,70,100,150,200,250]
for i in range(len(x)):
    temp_cpu=[[]for z in range(len(l))]
    temp_memory=[[]for z in range(len(l))]
    sys_cpu_temp=[]
    with open("data/raspberry-pi/"+str(x[i])+".txt","r") as file:
        for line in file:
            if "%Cpu(s):" in line:
                sys_cpu_temp.append(100-float(line.split()[7]))
            for j in range(len(l)):
                if l[j] in line:
                    temp=line.split()
                    temp_cpu[j].append(float(temp[8]))
                    temp_memory[j].append(float(temp[9]))
    sys_cpu.append(sum(sys_cpu_temp)/len(sys_cpu_temp))
    for j in range(len(l)):
        if (len(temp_cpu[j])==0):
            cpu[j].append(0)
            memory[j].append(0)
        else:
            cpu[j].append(sum(temp_cpu[j])/(4*len(temp_cpu[j])) )
            memory[j].append(sum(temp_memory[j])/len(temp_memory[j]) )
```

```python
def list_add(*a):
    # print(a)
    c = [0]*len(a[0])
    for i in a:
        for j in range(len(i)):
            c[j]=c[j]+i[j]
    return c
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
width = 0.6     
ind = np.arange(len(x))
print(ind)
font1 = {'family' : 'Arial',
'weight' : 'normal',
'size'   : 25,
}


font2 = {'family' : 'Arial',
'weight' : 'normal',
'size'   : 15,
}


font3 = {'family' : 'Arial',
'weight' : 'normal',
'size'   : 22,
}

mpl.rcParams['hatch.linewidth'] = '0.5'
mpl.rcParams['hatch.color'] = 'black'
plt.figure(figsize=(6, 4))
plt.ylim(0, 100)

p8 = plt.bar(ind, sys_cpu, width,edgecolor='black')
p1 = plt.bar(ind, cpu[0], width,edgecolor='black')
p2 = plt.bar(ind, cpu[1], width,edgecolor='black',
             bottom=cpu[0])
p3 = plt.bar(ind, cpu[2], width,edgecolor='black',
             bottom=list_add(cpu[1],cpu[0]))
p4 = plt.bar(ind, cpu[4], width,edgecolor='black',
             bottom=list_add(cpu[1],cpu[0],cpu[2]))
p5 = plt.bar(ind, list_add(cpu[3],cpu[5]), width,edgecolor='black',
             bottom=list_add(cpu[1],cpu[0],cpu[2],cpu[4]))
p6 = plt.bar(ind, cpu[6], width,edgecolor='black',
             bottom=list_add(cpu[1],cpu[0],cpu[2],cpu[4],cpu[3],cpu[5]))
p7 = plt.bar(ind, cpu[-1], width,edgecolor='black',
             bottom=list_add(cpu[1],cpu[0],cpu[2],cpu[4],cpu[3],cpu[5]))
plt.xlabel("Initial/Mobility registrations per second",font3)
plt.ylabel("Satellite CPU(%)",font1)

plt.tick_params(labelsize=15)
plt.xticks(ind, x)
legend1=plt.legend((p1[0], p2[0],p3[0],p4[0]), ('UPF', 'SMF','AMF','PCF'),prop=font2,ncol=2,frameon=False,bbox_to_anchor=(0.58, 0.75))
legend2=plt.legend((p5[0],p8[0],p6[0],p7[0]),('UDM','Others','AUSF','DB'),prop=font2,frameon=False,bbox_to_anchor=(0.328, 0.35))
plt.gca().add_artist(legend1)
plt.gca().add_artist(legend2)

plt.tight_layout()
plt.savefig("figure7a.pdf")

plt.show()
```
