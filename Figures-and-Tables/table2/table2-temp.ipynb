```python
#conding=utf8  
import os 
from datetime import datetime 
#mian data folder
data_dir_explore_710="../../Dataset/satellite-terminal-data/explore-710/"
data_dir_sc310="../../Dataset/satellite-terminal-data/tiantong-sc310/"
data_dir_T900="../../Dataset/satellite-terminal-data/tiantong-T900/"
```

```python
g = os.walk(data_dir_explore_710)  
all_lines=0
sm_lines=0
mm_lines=0
rrc_lines=0
l1_l2_lines=0
dirname=""
i=1
for path,dir_list,file_list in g:  
    for file_name in file_list:  
        filepath=os.path.join(path, file_name)
        if "RM_diagnostic" in filepath:
            try:
                processed_lines=[]
                with open(filepath) as f:
                    lines=f.readlines()
                    all_lines+=len(lines)
                    flag=0
                    for line in lines:
                        if "UMTS-SM" in line:
                            sm_lines+=1
                            processed_lines.append("Session management")
                            flag=1
                        if "UMTS-MM" in line or "UMTS-GMM" in line:
                            mm_lines+=1
                            processed_lines.append("Mobility management")
                            flag=1
                        if "IAI2:BCT" in line or "IAI2:BCN" in line:
                            rrc_lines+=1
                            processed_lines.append("RRC layer")
                            flag=1
                        if "SAT_SEARCH" in line:
                            l1_l2_lines+=1
                            processed_lines.append("L1/L2")
                            flag=1
                        if flag==0:
                            processed_lines.append("Others")
                        flag=0
                f.close()
                if dirname!=filepath.split("/")[-3]:
                    i=1
                    dirname=filepath.split("/")[-3]
                else:
                    i+=1
                dirpath="data/explore-710/"+dirname
                filename_after="data/explore-710/"+dirname+"/"+str(i)+".txt"
                print(filename_after)
                if not os.path.exists(dirpath):
                    os.makedirs(dirpath) 
                with open(filename_after,"w") as f:
                    for line in processed_lines:
                        f.write(line+"\n")
                    f.close()
            except Exception as e:
                print(e)
print("Explore 710 data lines:")
print("L1/L2: {}\nRRC: {}\nMM: {}\nSM: {}\nTotal: {}\n".format(l1_l2_lines,rrc_lines,mm_lines,sm_lines,all_lines))

```

```python
g = os.walk(data_dir_sc310)  
all_lines=0
sm_lines=0
mm_lines=0
rrc_lines=0
l1_l2_lines=0
dirname=""
i=1
for path,dir_list,file_list in g:  
    for file_name in file_list:  
        filepath=os.path.join(path, file_name)
        if "data-plane" not in filepath:
            try:
                print(filepath)
                processed_lines=[]
                with open(filepath) as f:
                    lines=f.readlines()
                    all_lines+=len(lines)
                    flag=0
                    for line in lines:
                        temp=line.lower()
                        if "sm" in temp and "SMAC" not in line and "reasm" not in temp:
                            sm_lines+=1
                            processed_lines.append("Session management")
                            flag=1                            
                        if "mm" in temp:
                            processed_lines.append("Mobility management")
                            flag=1                            
                            mm_lines+=1
                        if "rrc" in temp :
                            processed_lines.append("RRC layer")
                            flag=1
                            rrc_lines+=1
                        if "l1" in temp or "SMAC" in line:
                            l1_l2_lines+=1
                            processed_lines.append("L1/L2")
                            flag=1
                        if flag==0:
                            processed_lines.append("Others")     
                        flag=0                    
                f.close()
                if dirname!=filepath.split("/")[-2]:
                    i=1
                    dirname=filepath.split("/")[-2]
                else:
                    i+=1
                dirpath="data/tiantong-sc310/"+dirname
                filename_after="data/tiantong-sc310/"+dirname+"/"+str(i)+".txt"
                print(filename_after)
                if not os.path.exists(dirpath):
                    os.makedirs(dirpath) 
                with open(filename_after,"w") as f:
                    for line in processed_lines:
                        f.write(line+"\n")
                    f.close()   
            except:
                pass
print("Tiantong SC310 data lines:")
print("L1/L2: {}\nRRC: {}\nMM: {}\nSM: {}\nTotal: {}\n".format(l1_l2_lines,rrc_lines,mm_lines,sm_lines,all_lines))

```
